## Question 1: The Peachy Peach Juice Stand
The Peachy Peach juice stand offers fresh peach juice. It charges $3.09 per glass for the first five glasses, $2.45 per glass for the next ten glasses, and $1.91 for each additional glass.

Write a program that asks the user for the number of juice glasses ordered and calculates the total cost of the order, displaying the breakdown of the calculation in both dollars and cents. The program should also show the amount for card payment (exact amount) and cash payment (rounded to the nearest 5 cents <u>closest</u>), always <u>displaying two decimal places</u>.

Please note that due to the limited size of the stand, it is not possible to order more than 24 glasses of juice per order. Orders exceeding this quantity, as well as orders of 0 or fewer glasses, will not be accepted, and the following message will be displayed to the user:

> "Please ensure you order at least 1 juice and no more than 24 juices."

```python
# Code your solution here.

def calculate_total_cost(num_glasses):

    # Establish three variables for the number of glasses

    num_glasses1 = 0
    num_glasses2 = 0
    num_glasses3 = 0
        

    # Define the number of glasses per price range
    if num_glasses <= 5:
        num_glasses1 = num_glasses
    elif num_glasses <= 15:
        num_glasses1 = 5
        num_glasses2 = num_glasses - 5
    else:
        num_glasses1 = 5
        num_glasses2 = 10
        num_glasses3 = num_glasses - 15

    # Display the calculation details
    print("\nInvoice details:")

    # Establish three variables for the price per range of glasses.
    subtotal1 = num_glasses1 * 3.09
    subtotal2 = num_glasses2 * 2.45
    subtotal3 = num_glasses3 * 1.91

    # Display the prices per range of glass quantities
    print("   {num_glasses1} glasses at $3.09 = ${subtotal1:.2f}".format(num_glasses1 = num_glasses1, subtotal1 = subtotal1))
    if num_glasses2 > 0:
        print("   {num_glasses2} glasses at $2.45 = ${subtotal2:.2f}".format(num_glasses2 = num_glasses2, subtotal2 = subtotal2))
    if num_glasses3 > 0:
        print("   {num_glasses3} glasses at $1.91 = ${subtotal3:.2f}".format(num_glasses3 = num_glasses3, subtotal3 = subtotal3))
    total_cost = subtotal1 + subtotal2 + subtotal3
    
    # Display the amount for card payment
    print("\nTotal (card): ${total_cost:.2f}".format(total_cost = total_cost))

    # Display the amount for cash payment (rounded to the nearest 5 cents)
    cash_amount = round(total_cost * 20) / 20  
    print("Total (cash): ${cash_amount:.2f}".format(cash_amount = cash_amount))

# Ask the user for the number of glasses ordered
try:
    num_glasses = int(input("How many glasses of juice would you like to order? ")) 
    
    # Validate the number of glasses ordered
    print("How many glasses of juice would you like to order? {num_glasses}".format(num_glasses = num_glasses))

    if num_glasses <= 0 or num_glasses > 24:
        print("\nPlease ensure you order at least 1 juice and no more than 24 juices.")
    else:
        calculate_total_cost(num_glasses)    
except ValueError:
    print("Please enter a valid integer.")
```

Develop a function named `creation_tree`. This function is intended to create Christmas tree logos using ASCII characters. The branches of the tree should be represented by the `*` character, and the trunk by `|`. An essential aspect of your task is to ensure that each tree is centered vertically (i.e., the trunk should be in the middle of the second-to-last line) for optimal aesthetic appearance. To achieve this, it is crucial that the number of branches per line is always odd.

<br> 
<u>Function Specifications:</u>

- Function Name: `creation_tree`
- Parameter:
    - <kbd>height</kbd>: This parameter, with a default value of 5, determines the number of branch lines for the tree.

<u>Your Task:</u>

### Analysis Context

You are a data analyst at LuxHotels, an international luxury hotel chain. You have been provided with two datasets, `reservations.csv` and `agents.csv`, to conduct an in-depth analysis. Your task is to extract relevant insights, focusing on customer behaviors, their reservation consumption habits, as well as the efficiency of travel agents.

### Data Description

`reservations.csv`: This file contains detailed information on reservations made at LuxHotels properties. Itâ€™s important to note that each *observation* (row) in the dataset corresponds to *one reservation*.

Details of variables in the `reservations.csv` file:  
1. `country`: customer's country of origin.
2. `children`: number of children accompanying the customer.
3. `stays_in_weekend_nights`: number of nights stayed during the weekend.
4. `stays_in_week_nights`: number of nights stayed during the week.
5. `daily_rate`: daily rate (i.e., amount paid per night).
6. `total_of_special_requests`: total number of special requests made by the customer.
7. `previous_cancellations`: number of prior reservations canceled by the customer.
8. `is_repeated_guest`: indicates if the customer is a returning guest (0 = no, 1 = yes).
9. `name`: full name of the customer.
10. `email`: customer's email address.
11. `arrival_date_year`: year of arrival.
12. `arrival_date_month`: month of arrival.
13. `agent_id`: travel agent identifier.
14. `phone_number`: customer's phone number.

<br>

**agents.csv**: This file contains information about travel agents who made reservations on behalf of LuxHotels.

Details of variables in the `agents.csv` file:  
1. `agent_id`: unique identifier of the travel agent.
2. `travel_agency`: name of the travel agency.
3. `agent_firstname`: agent's first name.
4. `agent_lastname`: agent's last name.


```python
def creation_tree(height=5):

    if height % 2 == 0:
        height += 1  # Ensure an odd number of branches per line of the tree

    tree = ""
# Adjust the number of tree lines according to the specified height
    if height == 5:
        lines = height
    else:
        lines = height - 1

# Calculate the maximum number of branches per line
    max_branches = height * 2 - 1 

    for i in range(lines):
# Add spaces to center the tree vertically
        spaces = " " * ((height - i - 1))

# Add branches to the tree
        branches = "*" * (i * 2 + 1)

        tree += spaces + branches + " \n "

# Add the centered tree trunk
    trunk = "|"
    tree += trunk.center(max_branches) + "\n"

    return tree

# Calls to the creation_tree function with different heights and storing the results in tree variables.
tree_2 = creation_tree(2)
tree_5 = creation_tree()  # Default height = (5)
tree_6 = creation_tree(6)

# Displaying the trees
print("1. For height = 2: \n \n {tree_2}".format(tree_2=tree_2))
print("2. For default height (5): \n \n {tree_5}".format(tree_5=tree_5))
print("3. For height = 6: \n \n {tree_6}".format(tree_6=tree_6))
```

## Question 3

Read the files `reservations.csv` and `agents.csv` and store them in two DataFrames named `df_res` and `df_agents`, respectively.

Then, perform the following manipulations on `df_res`:

- Remove the `children` column from the dataset.
- Replace missing values in the `country` column with the most frequent value (i.e., the mode) in this column.
- Replace missing values in the `daily_rate` column with the median of this column.

---

```python

import pandas as pd
pd.__version__
```

```python
# Reading the CSV files
df_res = pd.read_csv('reservations.csv')
df_agents = pd.read_csv('agents.csv')

# Removing the 'children' column from df_res
df_res.drop(columns=['children'], inplace=True)

# Replacing missing values in the 'country' column with the mode
mode_country = df_res['country'].mode()[0]
df_res['country'].fillna(mode_country, inplace=True)

# Replacing missing values in the 'daily_rate' column with the median
median_daily_rate = df_res['daily_rate'].median()
df_res['daily_rate'].fillna(median_daily_rate, inplace=True)

# Displaying the first ten rows for verification
df_res.head(10)
```

## Question 4

1. From the DataFrame `df_res`, create a new column `tot_nights` that represents the total number of nights (week and weekend) for each reservation.

2. Create a new column `tot_cost` that calculates the total cost of the stay for each reservation, knowing that:

$$ \text{total cost} = \text{total number of nights} \times \text{daily rate} $$

```python

# 1. Create a new column tot_nights
df_res['tot_nights'] = df_res['stays_in_weekend_nights'] + df_res['stays_in_week_nights']

# 2. Create a new column tot_cost
df_res['tot_cost'] = df_res['tot_nights'] * df_res['daily_rate']

# Display the first ten rows for verification
df_res.head(10)
```

## Question 5

Filter `df_res` to isolate "demanding" customers.  
Customers are defined as "demanding" if they meet the following criteria for a given reservation:

1. Customers who made 3 or more special requests (`total_of_special_requests`).

2. Customers who have canceled (`previous_cancellations`) at least two previous reservations.

3. Customers who are not recurring guests (`is_repeated_guest`).

Store your result in a DataFrame named `df_filtered`, which will contain only the columns `name`, `email`, `daily_rate`, and `tot_cost`.


```python
# Filter demanding customers
df_filtered = df_res[(df_res['total_of_special_requests'] >= 3) & 
                     (df_res['previous_cancellations'] >= 2) & 
                     (df_res['is_repeated_guest'] == 0)]

# Select the required columns
df_filtered = df_filtered[['name', 'email', 'daily_rate', 'tot_cost']]

# Display rows for verification
df_filtered.head()
```

## Question 6

The `arrival_date_month` column in the DataFrame `df_res` indicates the full name of the month.

Create a new column named `mth` that will contain only the first three characters of the reservation month.

Delete the `arrival_date_month` column from the DataFrame `df_res` once the manipulation is completed.

```python
# Create a new column 'mth' containing the first three characters of the month
df_res['mth'] = df_res['arrival_date_month'].str[:3]

# Delete the 'arrival_date_month' column
df_res.drop(columns=['arrival_date_month'], inplace=True)

# Display the first ten rows for verification
df_res.head(10)
```

## Question 7

Each email address (`email`) in the DataFrame `df_res` contains information about the email service provider used by the customer.

Create a function to extract only the name of the email service provider from each email address in the `email` column.

Create a new column named `mail_company` that will store the name of the email service provider for each customer.

Once this operation is completed, analyze the frequency of use of the different email providers and display (`print`) the following message:

> The top 3 email providers from which the most reservations are made are: `<company_1>`, `<company_2>`, and `<company_3>` with `<count_1>`, `<count_2>`, and `<count_3>` reservations, respectively.

***Note***: To receive credit, the values of the requested variables cannot be *hard-coded*; you must use `.format()` or an *f-string* to print the variable values.


```python
def extract_mail_company(email):
    # Extract the name of the email service provider from each email address.
    return email.split('@')[-1].split('.')[0]

# Create a new column 'mail_company'
df_res['mail_company'] = df_res['email'].apply(extract_mail_company)

# Analyze the frequency of use of email service providers
company = df_res['mail_company'].value_counts().head(3)

# Print the message with the requested information
print("The top 3 email providers from which the most reservations are made are: "
      "{}, {} and {} with {}, {} and {} reservations, respectively.".format(
          company.index[0], company.index[1], company.index[2],
          company.iloc[0], company.iloc[1], company.iloc[2]
      ))
```

## Question 8

Create a new DataFrame named `df_join` by joining `df_res` to `df_agents`.

The resulting DataFrame should contain all the variables from the original DataFrames. However, we want to move the `mth` column to the right of the `arrival_date_year` column.

```python
# Join df_res to df_agents
df_join = df_res.merge(df_agents, on='agent_id')

# Move the 'mth' column to the right of the 'arrival_date_year' column
columns = list(df_join.columns)
mth_index = columns.index('mth')
year_index = columns.index('arrival_date_year')
columns.insert(year_index + 1, columns.pop(mth_index))
df_join = df_join[columns]

# Display the first ten rows for verification
df_join.head(10)
```

Use the `df_join` DataFrame to identify the 10 countries with the highest number of reservations.
